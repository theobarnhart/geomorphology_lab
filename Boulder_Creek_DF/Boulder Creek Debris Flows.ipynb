## Plot the velocity data vs the contributing area

```python
import statsmodels.formula.api as sm
```

```python
yolo_velo = [7.48,8.06,8.58,6.4,7.3,8]
yolo_area = pd.DataFrame([105125,105125,105125,46009,46009,46009], columns = ['area'])
yolo_area['const'] = np.repeat(1,6)
```

```python
yolo_area
```

```python
yolo = sm.OLS(yolo_velo,yolo_area)
```

```python
res = yolo.fit()
```

```python
plt.plot(yolo_area.area,yolo_velo,'ok')
plt.xlabel('Contributing Area [m$^2$]', fontsize = 18)
plt.xlim([35000,130000])
plt.ylim([4,10])
plt.plot(yolo_area,res.fittedvalues,'r-')
plt.title('Yolo Creek Velocity vs Contributing Area', fontsize = 24)
plt.ylabel('Velocity [m/s]', fontsize = 18)
plt.axvline(120233,c='k');
plt.savefig(outpath+'yolo_velo.pdf',bbox_in='tight')
```

```python
woof_velo = [4.8,5.3,5.7,4.7,5.7,6.5]
woof_area = pd.DataFrame([146343,146343,146343,82264,82264,82264], columns = ['area'])
woof_area['const'] = np.repeat(1,6)
```

```python
woof = sm.OLS(woof_velo,woof_area)
```

```python
res = woof.fit()
```

```python
plt.plot(woof_area.area,woof_velo,'ok')
plt.xlabel('Contributing Area [m$^2$]', fontsize = 18)
plt.xlim([35000,200000])
plt.ylim([4,8])
plt.plot(woof_area,res.fittedvalues,'r-')
plt.title('Woof Creek Velocity vs Contributing Area', fontsize = 24)
plt.ylabel('Velocity [m/s]', fontsize = 18)
plt.axvline(173372,c='k');
plt.savefig(outpath+'woof_velo.pdf',bbox_in='tight')
```
